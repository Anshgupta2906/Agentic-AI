HOW TO CALL A GOOGLE GEMINI MODEL

```python
import os
from langchain.chat_models import init_chat_model
os.environ["GEMINI_API_KEY"] = os.getenv("GEMINI_API_KEY")

model=init_chat_model("google_genai:gemini-2.5-flash-lite")
response=model.invoke("Why do parrot talk?")
response
```

Streaming

```python
for chunk in model.stream("Why do parrot have colourful feathers?"):
    print(chunk, end="|", flush=True)
```

BATCH


```python
responses=model.batch([
    "why do parrots talk?",
    "how do airplanes fly?",
    "what is cryptocurrency?",
    "how to become a billionaire in india?"

])
for response in responses:
    print(response)
```

```python
### using batch_as_completed to see if it streams? or not

responses=model.batch_as_completed(
    
    ["why do parrots talk?",
    "how do airplanes fly?",
    "what is cryptocurrency?",
    "how to become a billionaire in india?"]
)
for response in responses:
    print(response)
```
